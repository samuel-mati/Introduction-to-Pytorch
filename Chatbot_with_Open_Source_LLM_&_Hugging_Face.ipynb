**Step 2:** Import Dependencies from the transformer library

```python
from transformers import AutoTokenizer,AutoModelForSeq2SeqLM
```

**Step 3:** Choose the model to use

This is direct in the Hugging Face
[Hugging Face Models](https://huggingface.co/models)

We select "facebook/blenderbot-400M-distill". Its open source and fast.

```python
model_name="facebook/blenderbot-400M-distill"
```

**Step 4:** Fetch and Initialize the Model

```python
model=AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer=AutoTokenizer.from_pretrained(model_name)
```

**Step 5:** Chatbot Process

To start chatting;

Initialize and store our conversation history

Afterwords, we do the following for each iteration



1. Encode conversation history as astring
2. Fetch prompt from the user
3. Tokenize(optimise) prompt
4. Generate output from model using prompt and history
5. Decode output
6. Update conversation history




**Step 5a:** Kreping track of conversation history


```python
conversation_history=[]
```

**Step 5b:** Encoding the conversation History

During each interaction we pass our conversation history to the model along with our input for referencing the previous conversation

```python
history_string="\n".join(conversation_history)

history_string
```

**Step 5c:** Fetch Output fromthe User

```python
input_text='Hi, How are you?'
```

**Step 5d:** Tokenization of UserPrompt and Chat History

Tokens in NLP are individual units or elements that textor sentences are divided into. Tokenization is the procession of coverting tokens into numerical representations.

In NLP tasks, we often use the `encode_plus` method from the tokenizer object to perform tokenization

```python
inputs= tokenizer.encode_plus(history_string,input_text, return_tensors="pt")

inputs
```

In doing so we've created a **Python Dictionary** which contains special keywords that allow the model to properly reference its contents.

For more information visit [Pretrained Vocab Files Map](https://huggingface.co/transformers/v4.2.2/main_classes/tokenizer.html)

```python
tokenizer.pretrained_vocab_files_map
```

**Step 5e:** Generate Output from the Model

Since we have the inputsready,both paast and present inputs,,we can pass them to the model to generate a response. As per the [Documentation](https://huggingface.co/docs) we use `generate()` functions and pass the inputs as keyword arguments (kwargs)

```python
outputs=model.generate(**inputs)
outputs
```

**Step 5f:** Decode the Output

We may decode the output using tokenizer.decode()

This is the reverseof tokenization; it entails combining or merging individual tokens back to the original form,typically reconstructing the original text or sentence.

```python
response= tokenizer.decode(outputs[0],skip_special_tokens=True).strip()
response
```

**Step 5g**: Update Conversation History

Add both input and response to the conversation history

```python
conversation_history.append(input_text)
conversation_history.append(response)
conversation_history
```

**Step 6:** Repeat

```python
while True:
  #Create conversation History String
  history_string="\n".join(conversation_history)

  # Get the input from user
  input_text=input(">>")

  # Tokenize the input text and history
  inputs=tokenizer.encode_plus(history_string, input_text, return_tensors='pt')

  # Generate the response from the model
  outputs=model.generate(**inputs)

  # Decode the response
  response=tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

  #Add interaction to conversation history
  conversation_history.append(input_text)
  conversation_history.append(response)

```
